```python
import pandas as pd
import json
from math import log
```

```python
data = pd.read_excel("Supply_Chain_Data_0506.xlsx", sheet_name=None)
splc_1 = data["Suppliers"]
splc_2 = data["SOS"]
company_info = data["ALL COMPANY"]
```

```python
data_centrality = pd.read_excel("Centrality.xlsx")
centrality = {}

def centrality_data(info):
    d = []
    for k in ["Degree Centrality", "Closeness Centrality", "Betweenness Centrality", "Hub score", "Auth score"]:
        d.append(info[k])
    return d

for _, info in data_centrality.iterrows():
    centrality[info["Ticker"]] = centrality_data(info)

values = [[x[i] for x in centrality.values()] for i in range(5)]
for x in centrality.values():
    for i in range(5):
        x[i] = (x[i] - min(values[i])) / (max(values[i]) - min(values[i]))
for i in range(5):
    print(min(values[i]), max(values[i]))
```

```python
nodes = []
market_caps = [[], []]
roas = [[], []]

for _, info in company_info.iterrows():
    if info['Type'] == 'customer': continue
    
    node = {}
    node['ticker'] = info['Ticker']
    node['name'] = info['Company Name']
    node['country'] = info['Country']
    node['region'] = info['Region']
    node['type'] = info['Type']
    node['roa'] = info['ROA']
    node['sector'] = info['Sector']
    node['industry'] = info['Industry']
    node['num_supplier'] = str(info['# Supplier']).replace('.0', '')
    if node['ticker'] not in centrality:
        centrality[node['ticker']] = [0, 0, 0, 0, 0]
    node['centrality_degree'] = centrality[node['ticker']][0]
    node['centrality_closeness'] = centrality[node['ticker']][1]
    node['centrality_betweenness'] = centrality[node['ticker']][2]
    node['centrality_hub'] = centrality[node['ticker']][3]
    node['centrality_auth'] = centrality[node['ticker']][4]
    
    market_cap = info['Market Cap USD']
    if market_cap == '#N/A' or market_cap == '--' or market_cap == 0 or pd.isna(market_cap):
        node["size"] = 5
        node["market_cap"] = "--"
        node["market_cap_percentile"] = '--'
    else:
        if type(market_cap) == str: print(market_cap)
        node["size"] = max(log(market_cap) - 12, 5)
        if node['type'] == 'manufacturer':
            market_caps[0].append((market_cap, node))
        else:
            market_caps[1].append((market_cap, node))
        node["market_cap"] = market_cap
    node['score'] = 1.0
    
    if pd.isna(node['roa']) or node['roa'] == '--': 
        node['roa'] = '--'
    else: 
        if type(node['roa']) == str: print(node['roa'])
        if node['type'] == 'manufacturer':
            roas[0].append((node['roa'], node))
        else:
            roas[1].append((node['roa'], node))
    if pd.isna(node['sector']): node['sector'] = '--'
    if pd.isna(node['industry']): node['industry'] = '--'
        
    nodes.append(node)
    
nodes[:2]
```

```python
links = []

for _, info in splc_1.iterrows():
    link = {}
    link['source'] = info['Supplier Ticker']
    link['target'] = info['Central Ticker']
    link['type'] = 'layer1'
    
    account = info['Account As Type']
    if account == ' ' or pd.isna(account): 
        link['account'] = 'N/A'
    else:
        link['account'] = account
    
    relation_value = info["Relationship Value USD"]
    link["relation"] = relation_value
    link["score"] = max(log(relation_value + 1) / 5 - 1, 1)
    
    links.append(link)
    
for _, info in splc_2.iterrows():
    link = {}
    link['source'] = info['Supplier Ticker']
    link['target'] = info['Central Ticker']
    link['type'] = 'layer2'
    
    account = info['Account As Type']
    if account == ' ' or pd.isna(account): 
        link['account'] = 'N/A'
    else:
        link['account'] = account
    
    relation_value = info["Relationship Value USD"]
    link["relation"] = relation_value
    link["score"] = max(log(relation_value + 1) / 5 - 1, 1)
    
    links.append(link)  
```

```python
# Rank the nodes and links
market_caps_central = sorted(market_caps[0], key=lambda x: x[0], reverse=True)
market_caps_supplier = sorted(market_caps[1], key=lambda x: x[0], reverse=True)
roas_central = sorted(roas[0], key=lambda x: x[0], reverse=True)
roas_supplier = sorted(roas[1], key=lambda x: x[0], reverse=True)

for i, (_, n) in enumerate(market_caps_central):
    n["market_cap_percentile"] = int(i / len(market_caps_central) * 100)
for i, (_, n) in enumerate(market_caps_supplier):
    n["market_cap_percentile"] = int(i / len(market_caps_supplier) * 100)

    
for i, (_, n) in enumerate(roas_central):
    n["roa_percentile"] = int(i / len(roas_central) * 100)
for i, (_, n) in enumerate(roas_supplier):
    n["roa_percentile"] = int(i / len(roas_supplier) * 100)
```

```python
print('# of nodes: ', len(nodes), '\n# of links: ', len(links))
```

```python
json.dump({"nodes": nodes, "links": links}, open("automobile_new.json", "w"), indent=2)
```
