```python
import pandas as pd
import json
from math import log, isnan
```

```python
data = pd.read_excel("./Bloomberg_Supply_Chain.xlsx", sheet_name=None)
main_data = data["SPLC_L2"]
main_data1 = data["SPLC_L1"]
company_info = data["Company Info"]
```

```python
company_cap = {}
for _, info in company_info.iterrows():
    if (info["Ticker"] not in company_cap) or company_cap[info["Ticker"]] == 0:
        company_cap[info["Ticker"]] = int(info["Market Cap USD"])
```

```python
def newnode(name, type, country, market_cap):
    if pd.isna(country): country = "--"
    
    node = {}
    if market_cap is None:
        node["size"] = 5
        node["market_cap"] = "--"
    else:
        node["size"] = max(log(market_cap) - 12, 5)
        node["market_cap"] = "USD {:,}".format(market_cap)
    node["name"] = name
    node["type"] = type
    node["score"] = 1.0
    node["country"] = country
    return node

nodes = []
company_id = {}
'''
for _, info in main_data1.iterrows():
    
    central = info["Central Company"]
    central_stock = info["Central Ticker"]
    central_country = info["Central Country"]
    
    if central_stock not in company_cap:
        company_cap[central_stock] = None
    
    if central_stock not in company_id:
        nodes.append(newnode(central, "central", central_country, company_cap[central_stock]))
        company_id[central_stock] = len(company_id)
        
        
for _, info in main_data1.iterrows():
    supplier = info["Supplier Company"]
    supplier_stock = info["Supplier Ticker"]
    supplier_country = info["Supplier Country"]
    
    if supplier_stock not in company_cap:
        company_cap[supplier_stock] = None
    
    if supplier_stock not in company_id:
        nodes.append(newnode(supplier, "supplier", supplier_country, company_cap[supplier_stock]))
        company_id[supplier_stock] = len(company_id)
'''

for _, info in main_data.iterrows():
    
    central = info["Central Company"]
    central_stock = info["Central Ticker"]
    central_country = info["Central Country"]
    
    if central_stock not in company_cap:
        company_cap[central_stock] = None
    
    if central_stock not in company_id:
        nodes.append(newnode(central, "central", central_country, company_cap[central_stock]))
        company_id[central_stock] = len(company_id)
        
        
for _, info in main_data.iterrows():
    supplier = info["Supplier Company"]
    supplier_stock = info["Supplier Ticker"]
    supplier_country = info["Supplier Country"]
    
    if supplier_stock not in company_cap:
        company_cap[supplier_stock] = None
    
    if supplier_stock not in company_id:
        nodes.append(newnode(supplier, "supplier", supplier_country, company_cap[supplier_stock]))
        company_id[supplier_stock] = len(company_id)

nodes[:5]
```

```python
json.dump({"nodes": nodes, "links": links}, open("automobile_sos.json", "w"), indent=2)
```
